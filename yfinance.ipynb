```python
msft = yf.Ticker("MSFT")
```

```python
# get all stock info
msft.info
```

1. **period** : período de dados para download (use o parâmetro period ou use start e end) Os períodos válidos são:
*“ 1d”, “5d”, “1mo”, “3mo”, “6mo”, “1y”, “2y”, “5y”, “10y”, “ytd”, “max”*

2. **interval** : intervalo de dados (dados de 1m estão disponíveis apenas para os últimos 7 dias e intervalo de dados <1d para os últimos 60 dias) Os intervalos válidos são:
*“ 1m”, “2m”, “5m”, “15m”, “30m”, “60m”, “90m”, “1h”, “1d”, “5d”, “1sem”, “1m”, “3m”*

3. **start** : Se não estiver usando ponto – no formato (aaaa-mm-dd) ou datahora.

4. **end** : Se não estiver usando ponto – no formato (aaaa-mm-dd) ou datahora.

5. **prepost** : Incluir dados de mercado pré e pós regulares nos resultados? (O padrão é  False)- geralmente não há necessidade de alterar isso de Falso

6. **auto_adjust** : Ajustar todos os preços OHLC (Open/High/Low/Close) automaticamente? (O padrão é  True) - apenas deixe isso sempre como verdadeiro e não se preocupe com isso

7. **actions** : Baixar eventos de dividendos e desdobramentos de ações? (O padrão é  True)
